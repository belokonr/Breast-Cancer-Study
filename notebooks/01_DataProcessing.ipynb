```python
import pandas as pd
```

```python
#load data
RF_FILE_1 = 'bcsc_risk_factors_summarized1_092020.csv'
RF_FILE_2 = 'bcsc_risk_factors_summarized2_092020.csv'
RF_FILE_3 = 'bcsc_risk_factors_summarized3_092020.csv'
MAM_FILE = 'BCSC_Mammography_Screening_Performance_Dataset.csv'
```

```python
#clean data
rf_1_df = pd.read_csv(RF_FILE_1)
rf_2_df = pd.read_csv(RF_FILE_2)
rf_3_df = pd.read_csv(RF_FILE_3)
mam_db = pd.read_csv(MAM_FILE)
```

```python
#for risk factor datasets, combine all three csv's using concatenate
rf_list = []
rf_list.append(rf_1_df)
rf_list.append(rf_2_df)
rf_list.append(rf_3_df)
rf_db = pd.concat(rf_list, ignore_index=True)
```

```python
rf_db.head()
```

```python
# remove all risk factor variables we aren't interested in
#rf_df = rf_db[['age_group_5_years', 'BIRADS_breast_density', 'current_hrt', 'bmi_group']]

cols_to_check = rf_db.columns.difference(['age_group_5_years', 'count'])

mask = (rf_db[cols_to_check] != 9).all(axis=1)

rf_df = rf_db[mask]

# remove all instances of unknown data and NaN's
#rf_df = rf_df.dropna()
#rf_df = rf_df[rf_df['BIRADS_breast_density'] != 9]
#rf_df = rf_df[rf_df['current_hrt'] != 9]
#rf_df = rf_df[rf_df['bmi_group'] != 9]

# transform age variable to store as 18-85 instead of 1-13
# the values in our case will be the lower limits of the age range for each group
rf_df['age_group_5_years'] = (rf_df['age_group_5_years'] * 5) + 20
# replace 25 with 18 as age group 1 is 18-29, unlike all others which go by 5 years
rf_df['age_group_5_years'].replace(25, 18, inplace=True)

# transform bmi variable to store as 10-35 instead of 1-4
# the values in our case will be the lower limits of the bmi range for each group
rf_df['bmi_group'] = (rf_df['bmi_group'] * 5) + 15
# replace 20 with 10 as bmi group 1 is 10-24.99, unlike all others which go by 5 units
rf_df['bmi_group'].replace(20, 10, inplace=True)

# name columns recognizeably
#rf_df.columns = ['Age', 'Breast Density', 'HRT', 'BMI']

rf_df.head()
```
